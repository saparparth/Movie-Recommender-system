```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')   #newlearn : this fuction helps in converting string of list into list
```

```python
def convert(obj):
    lisy=[]
    for i in ast.literal_eval(obj):
#         print(obj)  
         lisy.append(i['name'])
    return lisy      
```

```python
movies['genres']=movies['genres'].apply(convert)  #value sent into convert function above and storing it in movie genres
```

```python
def convert_cast(obj): #function for findind main crew members in aboive list
    lisy=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            lisy.append(i['name'])
            counter+=1
        else:
            break
    return lisy       
```

```python
movies['cast']=movies['cast'].apply(convert_cast) #putting cast members in df
```

```python
movies['crew'][0]
```

```python
def fetch_direct(obj):     #findind function for director in directory ablove for all columns
    lisy=[]
#     counter=0
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            lisy.append(i['name'])
            break
    return lisy       
```

```python
movies['crew']=movies['crew'].apply(fetch_direct) 
```

```python
ndf['tags']=ndf['tags'].apply(lambda x:x.lower())
```

```python
# steaming is tecnique it convert same words that is used in diffrent concepts into one word
# like ['loved','loving','love'] into ["love","love","love"]
# WELL BE USING NLTK LIBREYEY FOR THIS 
# NATURAL LAGUAGE PROCESSING LIBREYRY
```

```python
import nltk
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
print(ps.stem('loved'))
```

```python
def ste(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
ndf['tags']=ndf['tags'].apply(ste)
```

```python
# Applying vectorization:: so to make this system we have to make vectores of tags and fetch closest to fetch vector to recommend 
# well be converting texts into vector  -> called as texte vectorization technique well be using is bag of wordss0o
```

```python
# strtergy :: 1- we are taking tags evaluating them for common words
#             2- extracting commond words 
#             3-now taking tags again and seeing how many times those common words repeats 
#             4 - making list of them how many times it repeats in that tag 
#             5- we have vector so now this will help us in recommder system well fetch closest
#               vector which is user wtaching to and recommend it  
#             6- and will not be conserdering stop words in here like "a , the  and , if"
```

```python
# Applying vectorization:
from sklearn.feature_extraction.text import CountVectorizer  
```

```python
cv =CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(ndf['tags']).toarray()
```

```python
# now we will calculate distance between movies less the distance is the similar the movie is 
# grater the distance lesser the similariteies
# for distance will calculate angle co-sine vector : theta between thr vector anglr is lesser / theets is lesser the lesser is the angle
# distance is inverly proprtional to similarity more the distance lesser the similarity
# well use co-sine similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors) #finding=g similarity between vectors angle  
```

```python
# making recommder system:

def recommend(movie):
    movie_idx=ndf['title'][ndf['title']==movie].index[0] #matching the title to index 
    distances=similarity[movie_idx] #finding distances of given movie in verctor simmilrization we created
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] # sorting the less distces top 5 movies and recommending it
    for i in movie_list:
        try:
            print(ndf.iloc[i[0]].title) #recommending the movie
        except:
            print("Movie Not found")
```

```python
recommend('1982')
```
